Minimum Path Sum

Given a m x n grid filled with non-negative numbers, find a path from top left to bottom right, which minimizes the sum of all numbers along its path.

Note: You can only move either down or right at any point in time.

 

Example 1:
```
Input: grid = [[1,3,1],[1,5,1],[4,2,1]]
Output: 7
```
Explanation: Because the path 1 → 3 → 1 → 1 → 1 minimizes the sum.

Example 2:
```
Input: grid = [[1,2,3],[4,5,6]]
Output: 12
```
 

Constraints:
```
m == grid.length
n == grid[i].length
1 <= m, n <= 200
0 <= grid[i][j] <= 200
```

````python
from typing import List

class Solution:
    def calculate(self, grid: List[List[int]], i: int, j: int) -> int:
        if i == len(grid) or j == len(grid[0]):
            return float("inf")
        if i == len(grid) - 1 and j == len(grid[0]) - 1:
            return grid[i][j]
        return grid[i][j] + min(
            self.calculate(grid, i + 1, j), self.calculate(grid, i, j + 1)
        )

    def minPathSum(self, grid: List[List[int]]) -> int:
        return self.calculate(grid, 0, 0)
    
grid = [[1,3,1],[1,5,1],[4,2,1]]
sol = Solution()
sol.minPathSum(grid)
````

````python
class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        memo = {}
        
        def calculate(i: int, j: int) -> int:
            if i == len(grid) or j == len(grid[0]):
                return float("inf")
            if i == len(grid) - 1 and j == len(grid[0]) - 1:
                return grid[i][j]
            if (i, j) not in memo:
                # Step 1: Calculate path going down
                down_path = calculate(i + 1, j)
                
                # Step 2: Calculate path going right
                right_path = calculate(i, j + 1)
                
                memo[(i, j)] = grid[i][j] + min(down_path, right_path)
            return memo[(i, j)]
        
        return calculate(0, 0)
    
grid = [[1,3,1],[1,5,1],[4,2,1]]
sol = Solution()
sol.minPathSum(grid)
````
