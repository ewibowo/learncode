Intersection of Two Linked Lists

Solution
Given the heads of two singly linked-lists headA and headB, return the node at which the two lists intersect. If the two linked lists have no intersection at all, return null.

The test cases are generated such that there are no cycles anywhere in the entire linked structure.

Note that the linked lists must retain their original structure after the function returns.

Custom Judge:

The inputs to the judge are given as follows (your program is not given these inputs):

* intersectVal - The value of the node where the intersection occurs. This is 0 if there is no intersected node.
* listA - The first linked list.
* listB - The second linked list.
* skipA - The number of nodes to skip ahead in listA (starting from the head) to get to the intersected node.
* skipB - The number of nodes to skip ahead in listB (starting from the head) to get to the intersected node.

The judge will then create the linked structure based on these inputs and pass the two heads, headA and headB to your program. If you correctly return the intersected node, then your solution will be accepted.
 
Example 1:
```
Input: intersectVal = 8, listA = [4,1,8,4,5], listB = [5,6,1,8,4,5], skipA = 2, skipB = 3

```
Explanation: The intersected node's value is 8 (note that this must not be 0 if the two lists intersect).
From the head of A, it reads as [4,1,8,4,5]. From the head of B, it reads as [5,6,1,8,4,5]. There are 2 nodes before the intersected node in A; There are 3 nodes before the intersected node in B.
- Note that the intersected node's value is not 1 because the nodes with value 1 in A and B (2nd node in A and 3rd node in B) are different node references. In other words, they point to two different locations in memory, while the nodes with value 8 in A and B (3rd node in A and 4th node in B) point to the same location in memory.

Constraints:
```
The number of nodes of listA is in the m.
The number of nodes of listB is in the n.
1 <= m, n <= 3 * 104
1 <= Node.val <= 105
0 <= skipA <= m
0 <= skipB <= n
intersectVal is 0 if listA and listB do not intersect.
intersectVal == listA[skipA] == listB[skipB] if listA and listB intersect.
``` 

Follow up: Could you write a solution that runs in O(m + n) time and use only O(1) memory?

````python
from typing import Optional

class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        visited = set()
        
        # Store all nodes from list A
        current = headA
        while current:
            visited.add(current)
            current = current.next
        
        # Check each node in list B
        current = headB
        while current:
            if current in visited:
                return current
            current = current.next
        
        return None
````

````python
# Helper Methods for Testing
def create_intersecting_lists(listA_vals, listB_vals, skipA, skipB):
    """
    Creates two intersecting linked lists based on the problem format.
    
    Args:
        listA_vals: List of values for list A
        listB_vals: List of values for list B
        skipA: Number of nodes before intersection in list A
        skipB: Number of nodes before intersection in list B
    
    Returns:
        Tuple of (headA, headB, intersection_node)
    """
    if skipA >= len(listA_vals) or skipB >= len(listB_vals):
        # No intersection case
        headA = create_linked_list(listA_vals)
        headB = create_linked_list(listB_vals)
        return headA, headB, None
    
    # Create the intersection part (shared nodes)
    intersection_vals = listA_vals[skipA:]
    intersection_head = create_linked_list(intersection_vals)
    
    # Create list A's unique part and connect to intersection
    if skipA == 0:
        headA = intersection_head
    else:
        headA = create_linked_list(listA_vals[:skipA])
        # Connect to intersection
        current = headA
        while current.next:
            current = current.next
        current.next = intersection_head
    
    # Create list B's unique part and connect to intersection
    if skipB == 0:
        headB = intersection_head
    else:
        headB = create_linked_list(listB_vals[:skipB])
        # Connect to intersection
        current = headB
        while current.next:
            current = current.next
        current.next = intersection_head
    
    return headA, headB, intersection_head


def create_linked_list(values):
    """Creates a linked list from a list of values."""
    if not values:
        return None
    
    head = ListNode(values[0])
    current = head
    
    for val in values[1:]:
        current.next = ListNode(val)
        current = current.next
    
    return head


def print_linked_list(head, name="List"):
    """Prints a linked list (careful: stops after 20 nodes to avoid infinite loops)."""
    values = []
    current = head
    count = 0
    max_nodes = 20
    
    while current and count < max_nodes:
        values.append(str(current.val))
        current = current.next
        count += 1
    
    if current:
        values.append("...")
    
    print(f"{name}: {' -> '.join(values)}")


def print_list_with_addresses(head, name="List"):
    """Prints a linked list with memory addresses to show intersection."""
    current = head
    count = 0
    max_nodes = 20
    
    print(f"\n{name}:")
    while current and count < max_nodes:
        print(f"  Node(val={current.val}, id={id(current)})")
        current = current.next
        count += 1


# Test Cases
def run_tests():
    solution = Solution()
    
    print("=" * 60)
    print("Test Case 1: Example with intersection at value 8")
    print("=" * 60)
    listA_vals = [4, 1, 8, 3, 9]
    listB_vals = [5, 6, 1, 8, 4, 5]
    skipA = 2
    skipB = 3
    
    headA, headB, expected_intersection = create_intersecting_lists(
        listA_vals, listB_vals, skipA, skipB
    )
    
    print(f"Input: listA = {listA_vals}, listB = {listB_vals}")
    print(f"       skipA = {skipA}, skipB = {skipB}")
    print()
    
    print_linked_list(headA, "List A")
    print_linked_list(headB, "List B")
    
    result = solution.getIntersectionNode(headA, headB)
    
    if result:
        print(f"\n✓ Intersection found at node with value: {result.val}")
        print(f"  Expected intersection value: {expected_intersection.val}")
        print(f"  Same node? {result is expected_intersection}")
    else:
        print("\n✗ No intersection found")
    
    # Show memory addresses to verify intersection
    print_list_with_addresses(headA, "List A (with addresses)")
    print_list_with_addresses(headB, "List B (with addresses)")

    print_linked_list(headA, "List A")
    print_linked_list(headB, "List B")

# Run the tests
if __name__ == "__main__":
    run_tests()
````
